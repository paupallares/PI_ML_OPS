* 1 . peliculas mes

```python
meses_es = {
    'January': 'enero',
    'February': 'febrero',
    'March': 'marzo',
    'April': 'abril',
    'May': 'mayo',
    'June': 'junio',
    'July': 'julio',
    'August': 'agosto',
    'September': 'septiembre',
    'October': 'octubre',
    'November': 'noviembre',
    'December': 'diciembre',
}

semana_es = {
    'Monday': 'lunes',
    'Tuesday': 'martes',
    'Wednesday': 'miercoles',
    'Thursday': 'jueves',
    'Friday': 'viernes',
    'Saturday': 'sabado',
    'Sunday': 'domingo',
  
}

df['month_release'] = df['release_date'].apply(lambda x: meses_es[x.strftime('%B')] if pd.notnull(x) else '')
df['day_of_week_release'] = df['release_date'].apply(lambda x: semana_es[x.strftime('%A')] if pd.notnull(x) else '')

#df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d')
#df['month_release'] = df['release_date'].apply(lambda x: meses_es[x.strftime('%B')])
#df['day_of_week_release'] = df['release_date'].apply(lambda x: semana_es[x.strftime('%A')])
```

```python
def peliculas_mes(mes:str):
    '''Se ingresa el mes y la funcion retorna la cantidad de peliculas que se
    estrenaron ese mes historicamente'''

    # Crear una variable con los meses de estreno
    df_mes = df[df['month_release'] == mes]

    # Contar cuantas veces aparece el mes en la columna
    cantidad = len(df_mes)
    
    return {'mes': mes, 'cantidad': cantidad}
```

```python
print(peliculas_mes('enero'))
print(peliculas_mes('febrero'))
print(peliculas_mes('marzo'))
print(peliculas_mes('abril'))
print(peliculas_mes('mayo'))
print(peliculas_mes('junio'))
print(peliculas_mes('julio'))
print(peliculas_mes('agosto'))
print(peliculas_mes('septiembre'))
print(peliculas_mes('octubre'))
print(peliculas_mes('noviembre'))
print(peliculas_mes('diciembre'))
```

* 2 . peliculas dia

```python
def peliculas_dia(dia):
    '''Se ingresa el dia y la funcion retorna la cantidad de peliculas que se
    estrebaron ese dia historicamente'''

    # Crear una nueva columna en el dataframe con los días de la semana de estreno
    df_dia = df[df['day_of_week_release'] == dia]

     # Contar cuantas veces aparece el día de la semana en la columna
    cantidad = len(df_dia)
    
    return {'dia': dia, 'cantidad': cantidad}
```

```python
print(peliculas_dia('lunes'))
print(peliculas_dia('martes'))
print(peliculas_dia('miercoles'))
print(peliculas_dia('jueves'))
print(peliculas_dia('viernes'))
print(peliculas_dia('sabado'))
print(peliculas_dia('domingo'))
```

```python
def franquicia(franquicia:str):
    '''Se ingresa la franquicia, retornando la cantidad de peliculas, 
    ganancia total y promedio'''

    # Filtrar las filas correspondientes a la franquicia específica
    franquicia_df = df[df['belongs_to_collection'] == franquicia]

    # Calcular la cantidad de películas
    cantidad = len(franquicia_df)

    # Calcular la ganancia total y promedio (mean)
    ganancia_total = franquicia_df['revenue'].sum()
    ganancia_promedio = franquicia_df['revenue'].mean() if cantidad > 0 else 0
    
    return {'franquicia': franquicia, 'cantidad': cantidad, 'ganancia_total': ganancia_total, 'ganancia_promedio': ganancia_promedio}
```

```python
def peliculas_pais(pais:str):
    '''Ingresas el pais, retornando la cantidad de peliculas producidas en el mismo'''

    pais_df = df[df['production_countries'].str.contains(pais, na=False)]

    cantidad = len(pais_df)
    
    return {'pais': pais, 'cantidad': cantidad}
```

```python

def productoras(productora:str):
    '''Ingresas la productora, retornando la ganancia toal y la cantidad de peliculas que
    produjeron'''

    productora_df = df[df['production_companies'].notna() & df['production_companies'].str.contains(productora)]
    
    ganancia_total = productora_df['revenue'].sum()
    cantidad = len(productora_df)

    return {'productora': productora, 'ganancia_total': ganancia_total, 'cantidad': cantidad}
```

* 6 . devolver retorno

```python
def retorno(pelicula:str):
    '''Ingresas la pelicula, retornando la inversion, la ganancia, el retorno y 
    el año en el que se lanzo'''

     # Filtrar las filas correspondientes a la película específica
    pelicula_df = df[df['title'] == pelicula]

     # Obtener el presupuesto, recaudación y año de estreno de la película
    inversion = pelicula_df['budget'].iloc[0]
    ganancia = pelicula_df['revenue'].iloc[0]
    retorno = pelicula_df['return'].iloc[0]
    anio = pelicula_df['release_date'].dt.strftime('%Y').iloc[0]

    return {'pelicula':pelicula, 'inversion':inversion, 'ganancia':ganancia,'retorno':retorno, 'anio':anio}
```

```python
print(retorno('Toy Story'))
print(retorno('Waiting to Exhale'))
```

* 7 . recomendacion simple por genero

```python
def recomendacion(titulo): 
    '''Ingresas un nombre de pelicula y te recomienda las similares en una lista de 5 valores
        como es una funcion basica, los agrupe por genero, 
        la funcion devuelve una lista con las 5 primeras peliculas del genero que encuentra
    ''' 
    # Filtrar las filas correspondientes a la película específica
    pelicula_fila = df[df['title'] == titulo]

    if pelicula_fila.empty:
        return {'error': 'La película no se encuentra en el DataFrame'}

    # Obtener el género de la película
    genero = pelicula_fila['genres'].iloc[0]

    # Filtrar las películas con el mismo género (excluyendo la película original)
    peliculas_relacionadas = df[df['genres'] == genero]['title'].tolist()
    peliculas_relacionadas.remove(titulo)  # Eliminar la película original de la lista

    # Tomar las primeras 5 películas de la lista
    peliculas_recomendadas = peliculas_relacionadas[:5]

    # Retornar los resultados en un diccionario
    return {'lista recomendada': peliculas_recomendadas}
```

```python
print(recomendacion('Toy Story'))
print(recomendacion('Waiting to Exhale'))
```

```python
# Llamar a la función recomendacion para obtener los resultados

resultado = recomendacion('Toy Story')
resultado
```
